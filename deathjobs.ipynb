```python
def pmr_compare(dis1, dis2):
    a = df[(df['Cause'] ==  dis1) & (df['Observed'] > 4) & (df['PMR_Adj'] > 100)].sort('PMR_Adj', ascending=False)
    b = df[(df['Cause'] ==  dis2) & (df['Observed'] > 4) & (df['PMR_Adj'] > 100)].sort('PMR_Adj', ascending=False)
    c = pd.merge(a, b, on='Occupation')
    result = c[['Occupation', 'PMR_Adj_x', 'PMR_Adj_y']]
    result.columns = ['Occupation', dis1, dis2]
    return result
```

```python
result1 = pmr_compare('Cancer of the pleura', 'Fibrosing alveolitis')
result2 = pmr_compare('Asbestosis', 'Cancer of the peritoneum')
result3 = pmr_compare('Asbestosis', 'Fibrosing alveolitis')
result4 = pmr_compare('Cancer of the peritoneum', 'Fibrosing alveolitis')
```
