```python
import requests
from StringIO import StringIO
from numpy import nan as NA
import pandas as pd
import zipfile
import re
%matplotlib inline
import matplotlib.pyplot as plt
import xlrd
```

```python
url = "http://www.ons.gov.uk/ons/rel/occupational-mortality/occupational-mortality-in-england-and-wales/occupational-mortality/occupational-mortality--men-age-65-74.zip"
r = requests.get(url)
z = zipfile.ZipFile(StringIO(r.content))
notes = pd.read_excel(z.open('Men65_74.xls'), sheetname=1)
df = pd.read_excel(z.open('Men65_74.xls'), sheetname=2)

```

```python
notes[15:]
```

```python
df.columns
```

```python
df.columns = [x.strip() for x in df.columns] 
```

```python
df.Cause.unique()
```

```python
df[df['Observed'] > 5].sort('PMR_Adj', ascending=False) # lets see the biggest proportional mortality ratio where observed deaths are > 5
```

```python
df[df['Observed'] > 5].sort('PMR_Adj', ascending=False).to_csv('BiggestPMRsForPaul.csv')
```
